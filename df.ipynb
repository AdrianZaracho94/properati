```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ar_properties.csv')
```

```python
df.head(5)
```

```python
# Filtro la variable precio para que me quede solamente los valores que sean numericos (ya que es una variable sucia)
```

```python
filtro = df['price'] > 1
df_filtrado = df[filtro]
print(df_filtrado)
```

```python
filtro = df['bathrooms'] > 1
df_filtrado = df_filtrado[filtro]
print(df_filtrado)
```

```python
df_filtrado.info()
```

```python
# como tenemos datos nan, lo que voy a hacer es reemplazarlos por cero
df_filtrado = df_filtrado.fillna(0)
```

```python
df_filtrado.info()
```

```python
df_filtrado
```

```python
df3 = df_filtrado[df_filtrado.l2.isin(['Capital Federal'])] #filtro para datos str para quedarme con capital
      
df3
```

```python
# comenzamos a renombrar las variables

df3 = df3.rename(columns={'ad_type':'borrar',
                        'start_date':'inicio_publi',
                        'end_date':'fin_publi',
                        'created_on':'borrar3',
                        'lat':'latitud', 
                        'lon':'longitud',
                        'l1':'pais',
                        'l2':'provincia',
                        'l3':'barrio', 
                        'l4':'barrio2', 
                        'l5':'borrar4', 
                        'l6':'borrar5', 
                        'rooms':'borrar6', 
                        'bedrooms':'borrar7', 
                        'bathrooms':'borrar8',
                        'surface_total':'superficie_total', 
                        'surface_covered':'superficie_cubierta', 
                        'price':'precio',
                        'currency':'moneda_publicada', 
                        'price_period':'periodo_precio',
                        'title':'titulo_publi', 
                        'description':'borrar9',
                        'property_type':'tipo_propiedad', 
                        'operation_type':'tipo_operacion'
                        })
```

```python
df3
```

```python
df4 = df3.drop(['borrar', 'borrar3','borrar4', 'borrar5',
                'borrar6', 'borrar7', 'borrar8','borrar9',
                'barrio2'],
                axis=1)
print(df4)
```

Cambio el tipo de dato de la variable precio para poder operarla como entero(int)

```python
df4[['precio']] = df4[['precio']].astype(int)
```

```python
df4.info()
```

```python
df4
```

```python
# En la variable barrios nos quedo algo sucia, ya que contempla zonas que no son barrios en si

```

```python
# GENERO DOS DF QUE SOLO CONTENGAN VENTAS O LOS DOS TIPOS DE ALQUILERES

ventas = df4[df4.tipo_operacion.isin(['Venta'])] #filtro para datos str para quedarme con ventas
alquileres = df4[df4.tipo_operacion.isin(['Alquiler','Alquiler temporal'])] #filtro para datos str para quedarme con alquileres

print(ventas)
print(alquileres)
```

```python
alquileres.info()
```

```python
ventas.info()
```

# Exportaré los datos armados para luego procesarlos en powerbi

```python
alquileres.to_excel('df_alquileres.xlsx',sheet_name='alquileres',index=False)
ventas.to_excel('df_ventas.xlsx',sheet_name='ventas',index=False)
df4.to_csv('base_total_f')
```
